```python
fb = 173
frs = [6,7,8,9]

path_shp = "/home/philipp/Data/edin_diss/GIS_wo/wo_2020_nutz.shp"
wo_geo = geopandas.read_file(path_shp)

wo_geo['fb'] = wo_geo['fb'].astype(np.uint8)
wo_geo['fr'] = wo_geo['fr'].astype(np.uint8)
wo_geo['abt'] = wo_geo['abt'].astype(np.uint16)
wo_geo['uabt'] = wo_geo['uabt'].astype(str)
wo_geo['teilfl'] = wo_geo['teilfl'].astype(np.uint8)

wo_geo['admin'] = wo_geo['admin'].astype(np.uint16)
wo_geo['length'] = wo_geo['length'].astype(np.float32)
wo_geo['area'] = wo_geo['area'].astype(np.float32)
wo_geo['fly_date'] = wo_geo['fly_date'].astype(str)
wo_geo['year_fly'] = wo_geo['year_fly'].astype(np.uint16)

wo_geo['WO'] = wo_geo['WO'].astype(str)

wo_geo['fb_sap'] = wo_geo['fb_sap'].astype(np.uint8)
wo_geo['fr_sap'] = wo_geo['fr_sap'].astype(np.uint8)
wo_geo['abt_sap'] = wo_geo['abt_sap'].astype(np.uint16)
wo_geo['uabt_sap'] = wo_geo['uabt_sap'].astype(str)
wo_geo['teilfl_sap'] = wo_geo['teilfl_sap'].astype(np.uint8)

wo_geo['start_term'] = wo_geo['start_term'].astype(str)
wo_geo['uz'] = wo_geo['uz'].astype(np.uint8)
wo_geo['non_forest'] = wo_geo['non_forest'].astype(np.uint8)
wo_geo['economy'] = wo_geo['economy'].astype(str)
wo_geo['ww_sw'] = wo_geo['ww_sw'].astype(str)
wo_geo['sw_type'] = wo_geo['sw_type'].astype(str)
wo_geo['sea_level'] = wo_geo['sea_level'].astype(np.uint16)
wo_geo['exp'] = wo_geo['exp'].astype(str)
wo_geo['slope'] = wo_geo['slope'].astype(np.uint8)
wo_geo['site_type'] = wo_geo['site_type'].astype(np.uint8)
wo_geo['veg_type'] = wo_geo['veg_type'].astype(str)
wo_geo['growth_are'] = wo_geo['growth_are'].astype(str)

wo_geo['age'] = wo_geo['age'].astype(np.uint16)
wo_geo['BL'] = wo_geo['BL'].astype(np.uint8)
wo_geo['FI'] = wo_geo['FI'].astype(np.uint8)
wo_geo['LH'] = wo_geo['LH'].astype(np.uint8)
wo_geo['NH'] = wo_geo['NH'].astype(np.uint8)
wo_geo['LA'] = wo_geo['LA'].astype(np.uint8)
wo_geo['KI'] = wo_geo['KI'].astype(np.uint8)
wo_geo['DG'] = wo_geo['DG'].astype(np.uint8)
wo_geo['SK'] = wo_geo['SK'].astype(np.uint8)

wo_geo['mass_ha'] = wo_geo['mass_ha'].astype(np.float32)
wo_geo['mass_ha_lh'] = wo_geo['mass_ha_lh'].astype(np.float32)
wo_geo['mass_ha_nh'] = wo_geo['mass_ha_nh'].astype(np.float32)
wo_geo['mass_tot'] = wo_geo['mass_tot'].astype(np.float32)
wo_geo['mass_tot_l'] = wo_geo['mass_tot_l'].astype(np.float32)
wo_geo['mass_tot_n'] = wo_geo['mass_tot_n'].astype(np.float32)

wo_geo['cut_area'] = wo_geo['cut_area'].astype(np.float32)
wo_geo['cut_lh'] = wo_geo['cut_lh'].astype(np.float32)
wo_geo['cut_nh'] = wo_geo['cut_nh'].astype(np.float32)
wo_geo['cut_sum'] = wo_geo['cut_sum'].astype(np.float32)

wo_geo['dr'] = wo_geo['dr'].astype(np.uint8)
wo_geo['bp'] = wo_geo['bp'].astype(np.uint8)
wo_geo['zp'] = wo_geo['zp'].astype(np.uint8)
wo_geo['ru'] = wo_geo['ru'].astype(np.uint8)
wo_geo['sg'] = wo_geo['sg'].astype(np.uint8)
wo_geo['ma'] = wo_geo['ma'].astype(str)
wo_geo['year_fe'] = wo_geo['year_fe'].astype(np.uint16)

wo_geo['age_diff'] = wo_geo['age_diff'].astype(np.uint8)
wo_geo['fly_y'] = wo_geo['fly_y'].astype(np.uint16)
wo_geo['fly_m'] = wo_geo['fly_m'].astype(np.uint8)
wo_geo['fly_d'] = wo_geo['fly_d'].astype(np.uint8)

wo_geo['cut_ma'] = wo_geo['year_fe'].astype(str)
wo_geo['cut_v_lh'] = wo_geo['cut_v_lh'].astype(np.uint16)
wo_geo['cut_v_nh'] = wo_geo['cut_v_nh'].astype(np.uint16)
wo_geo['cut_v_nan'] = wo_geo['cut_v_nan'].astype(np.uint16)
wo_geo['cut_v_tot'] = wo_geo['cut_v_tot'].astype(np.uint16)

# filter data
wo_geo = wo_geo.loc[(wo_geo['fb']==fb) & (wo_geo['fr'].isin(frs)), :]
```

```python
# conditions 1 = BL, 2 = LH, 3 = NH, 4 = rest
conditions = [
    (wo_geo['non_forest'] == 1) | (wo_geo['non_forest'] == 2) | \
    (wo_geo['non_forest'] == 3) | (wo_geo['non_forest'] == 6) | \
    (wo_geo['non_forest'] == 7) | (wo_geo['non_forest'] == 8) | \
    (wo_geo['non_forest'] == 9),
    ((wo_geo['age'] > 10) & (wo_geo['dr'] == 1)),# | (wo_geo['age'] > 10) & (wo_geo['dr'] == 2)) | ((wo_geo['age'] > 10) & (wo_geo['dr'] > 3)), 
    ((wo_geo['age'] > 10) & (wo_geo['dr'] == 2)) | ((wo_geo['age'] > 10) & (wo_geo['dr'] > 3)),
    (wo_geo['age'] <= 10) | (wo_geo['dr'] == 0) | (wo_geo['dr'] == 3) | (wo_geo['non_forest'] == 4) | (wo_geo['non_forest'] == 5),]

choices = [4, 1, 2, 3]

wo_geo['typ'] = np.select(conditions, choices, default=0)
```

```python
wo_geo['typ'] = wo_geo['typ'].astype(int)
```
