```python
path_shp = "/home/philipp/Data/edin_diss/GIS_ground_truth/df_2020_2/gis_df_wwie_2020_176.shp"
wo_geo = gpd.read_file(path_shp)
```

```python
wo_geo.columns
```

```python
# conditions 0 = void, 1 = thinning, 2 = no thinning, 3 = else
conditions = [
    (wo_geo['typ'] == 0),
    (wo_geo['typ'] == 1),
    (wo_geo['typ'] == 2),
    ((wo_geo['typ'] == 3) | (wo_geo['typ'] == 4))]

choices = [0, 1, 2, 3]

wo_geo['typ_std'] = np.select(conditions, choices, default=0)
```

```python
# conditions 0 = void, 1 = thinning, 2 = just thinned, 3 = no thinning, 4 = else
conditions = [
    (wo_geo['typ'] == 0),
    (wo_geo['typ'] == 1),
    (wo_geo['typ'] == 4),
    (wo_geo['typ'] == 2),
    (wo_geo['typ'] == 3)]

choices = [0, 1, 2, 3, 4]

wo_geo['typ_add'] = np.select(conditions, choices, default=0)
```

```python
# conditions 0 = void, 1 = thinning (dr1), 2 = thinning (dr2), 3 = no thinning, 4 = else
conditions = [
    (wo_geo['typ'] == 0),
    ((wo_geo['typ'] == 1) & (wo_geo['dr'] == 1)),
    ((wo_geo['typ'] == 1) & (wo_geo['dr'] != 1)),
    (wo_geo['typ'] == 2),
    (wo_geo['typ'] == 3)]

choices = [0, 1, 2, 3, 4]

wo_geo['typ_dr'] = np.select(conditions, choices, default=0)
```
