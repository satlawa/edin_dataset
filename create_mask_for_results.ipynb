```python
path_shp = "/home/philipp/Data/edin_diss/GIS_ground_truth/df_2020_2/gis_df_wwie_2020_177.shp"
wo_geo = gpd.read_file(path_shp)
```

```python
wo_geo.columns
```

```python
wo_geo['uz'] - wo_geo['age'] > 40
```

```python
# binary

# if WW iE

# conditions 0 = void, 1 = thinning, 2 = no thinning, 3 = else
conditions = [
    wo_geo['typ'] == 0,
    (((wo_geo['ww_sw'] == 'W') & (wo_geo['economy'] == 'I') & (wo_geo['uz'] - wo_geo['age'] >= 30)))]
    #((wo_geo['typ'] == 3) | ((wo_geo['ww_sw'] == 'W') & (wo_geo['economy'] == 'I') & (wo_geo['uz'] - wo_geo['age'] >= 30)))]

choices = [0, 1]

wo_geo['mask'] = np.select(conditions, choices, default=0)
```
